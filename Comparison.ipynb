```python
import matplotlib.pyplot as plt
import numpy as np

def combine_cnn_phases(custom_cnn_results):
    """
    Combines all Custom CNN phases into a single timeline.
    """
    combined = {
        "train_loss": [],
        "val_loss": [],
        "val_acc": [],
        "val_auc": []
    }
    for phase_name, phase_data in custom_cnn_results.items():
        epoch_metrics = phase_data["epoch_metrics"]
        for metric, values in epoch_metrics.items():
            combined[metric].extend(values)
    return combined

# Unified model dictionary
all_models = {
    "DenseNet121": densenet121_results,
    "ResNet50": resnet50_results,
    "EfficientNet-B4": efficientnet_b4_results,
    "CustomCNN": {"epoch_metrics": combine_cnn_phases(custom_cnn_results)},
}


def get_epoch_values(model_dict, metric_name):
    """
    Extract epochs and metric values from model data.
    Supports both list-of-dicts and dict-of-lists formats.
    Automatically maps 'test_loss' and 'val_loss' to the same metric.
    """
    epoch_data = model_dict.get("epoch_metrics", {})
    
    # Standardize 'test_loss' and 'val_loss'
    if metric_name == "val_loss":
        metric_name = "test_loss" if "test_loss" in epoch_data else "val_loss"

    if isinstance(epoch_data, list):
        epochs = [entry["epoch"] for entry in epoch_data if metric_name in entry]
        values = [entry[metric_name] for entry in epoch_data if metric_name in entry]
        return epochs, values
    elif isinstance(epoch_data, dict):
        values = epoch_data.get(metric_name, [])
        epochs = list(range(1, len(values) + 1))
        return epochs, values
    return [], []



def plot_metric(metric_name, all_models_dict):
    """
    Plots a specific metric for all models across epochs.
    """
    plt.figure(figsize=(12, 8))
    for name, model_data in all_models_dict.items():
        epochs, values = get_epoch_values(model_data, metric_name)
        if epochs and values:
            plt.plot(epochs, values, marker='o', label=name)
    plt.title(f"{metric_name.capitalize()} Across Epochs")
    plt.xlabel("Epoch")
    plt.ylabel(metric_name.capitalize())
    plt.legend()
    plt.grid(True)
    plt.show()


def get_final_metric(model_dict, metric_name):
    """
    Retrieves the final value of a given metric from the model dictionary.
    """
    epoch_data = model_dict.get("epoch_metrics", {})
    
    # Standardize 'test_loss' and 'val_loss'
    if metric_name == "val_loss":
        metric_name = "test_loss" if "test_loss" in epoch_data else "val_loss"

    if isinstance(epoch_data, list):
        return epoch_data[-1].get(metric_name, None)
    elif isinstance(epoch_data, dict):
        values = epoch_data.get(metric_name, [])
        return values[-1] if values else None
    return None



if __name__ == "__main__":
    metrics = ["train_loss", "val_loss", "val_acc", "val_auc"]
    for metric in metrics:
        plot_metric(metric, all_models)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

confusion_matrices = {
    "DenseNet121": densenet121_results.get("confusion_matrix"),
    "ResNet50": resnet50_results.get("confusion_matrix"),
    "EfficientNet-B4": efficientnet_b4_results.get("confusion_matrix"),
    "CustomCNN": custom_cnn_results["phase3_31_epochs"].get("confusion_matrix")
}


def normalize_confusion_matrix(cm):
    """
    Normalizes a confusion matrix by row (true label).
    """
    return cm / np.sum(cm, axis=1, keepdims=True)


def plot_confusion_matrices(models_cm):
    """
    Plots confusion matrices for each model side by side.
    """
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    for ax, (model, cm) in zip(axes, models_cm.items()):
        cm_normalized = normalize_confusion_matrix(np.array(cm))
        sns.heatmap(cm_normalized, annot=True, cmap='Blues', fmt=".2f", ax=ax, cbar=False)
        ax.set_title(model)
        ax.set_xlabel("Predicted Label")
        ax.set_ylabel("True Label")
    plt.tight_layout()
    plt.show()



if __name__ == "__main__":
    # A) Plot normalized confusion matrices side by side
    print("\n Plotting Normalized Confusion Matrices:")
    plot_confusion_matrices(confusion_matrices)
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example classification reports from your models
classification_reports = {
    "DenseNet121": densenet121_results.get("classification_report"),
    "ResNet50": resnet50_results.get("classification_report"),
    "EfficientNet-B4": efficientnet_b4_results.get("classification_report"),
    "CustomCNN": custom_cnn_results["phase3_31_epochs"].get("classification_report")
}

# Accuracy from val_acc
val_acc = {
    "DenseNet121": densenet121_results["epoch_metrics"][-1]["val_acc"],
    "ResNet50": resnet50_results["epoch_metrics"][-1]["val_acc"],
    "EfficientNet-B4": efficientnet_b4_results["epoch_metrics"][-1]["val_acc"],
    "CustomCNN": custom_cnn_results["phase3_31_epochs"]["epoch_metrics"]["val_acc"][-1],
}

#############################################
# 1) Extract Malignant Metrics
#############################################
def extract_malignant_metrics(report, model_name):
    """
    Extract precision, recall, f1-score for the Malignant class.
    Accuracy is taken from val_acc.
    """
    return {
        "precision": report["Malignant"]["precision"],
        "recall": report["Malignant"]["recall"],
        "f1_score": report["Malignant"]["f1_score"],
        "accuracy": val_acc[model_name]  # Override with val_acc
    }

# Standardize metrics for all models
malignant_metrics = {name: extract_malignant_metrics(report, name) for name, report in classification_reports.items()}

#############################################
# 2) Plot Malignant Metrics Across Models
#############################################
def plot_malignant_metrics(metrics_dict):
    """
    Plot precision, recall, f1-score, and accuracy for Malignant cases across all models.
    """
    metrics = ["precision", "recall", "f1_score", "accuracy"]
    values = {metric: [] for metric in metrics}
    
    for metric in metrics:
        for model in metrics_dict.keys():
            values[metric].append(metrics_dict[model][metric])
    
    x = np.arange(len(metrics))  # Metric indices
    width = 0.2  # Width of each bar
    
    plt.figure(figsize=(14, 8))
    for i, (model, color) in enumerate(zip(metrics_dict.keys(), ['skyblue', 'lightgreen', 'orange', 'pink'])):
        model_values = [metrics_dict[model][metric] for metric in metrics]
        plt.bar(x + i * width, model_values, width, label=model, color=color)
    
    plt.title("Malignant Metrics Across Models")
    plt.ylabel("Score")
    plt.xlabel("Metrics")
    plt.xticks(x + width, metrics)
    plt.legend(title="Model")
    plt.ylim(0.85, 1.0)  # Compact Y-axis
    
    # Annotate each bar
    for i, (model, color) in enumerate(zip(metrics_dict.keys(), ['skyblue', 'lightgreen', 'orange', 'pink'])):
        model_values = [metrics_dict[model][metric] for metric in metrics]
        for j, value in enumerate(model_values):
            plt.text(x[j] + i * width, value + 0.001, f"{value:.3f}", ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

#############################################
# 3) Plot Differences from Mean for Malignant Metrics
#############################################
def plot_malignant_differences(metrics_dict):
    """
    Plot differences from the mean for precision, recall, f1-score, and accuracy in Malignant cases.
    """
    metrics = ["precision", "recall", "f1_score", "accuracy"]
    differences = {model: [] for model in metrics_dict.keys()}
    mean_values = {metric: np.mean([metrics_dict[model][metric] for model in metrics_dict]) for metric in metrics}
    
    for model in metrics_dict.keys():
        for metric in metrics:
            differences[model].append(metrics_dict[model][metric] - mean_values[metric])
    
    x = np.arange(len(metrics))  # Metric indices
    width = 0.2  # Width of each bar
    
    plt.figure(figsize=(14, 8))
    for i, (model, color) in enumerate(zip(metrics_dict.keys(), ['skyblue', 'lightgreen', 'orange', 'pink'])):
        model_diffs = differences[model]
        plt.bar(x + i * width, model_diffs, width, label=model,
                color=['green' if diff > 0 else 'red' for diff in model_diffs])
    
    plt.title("Differences from Mean for Malignant Metrics Across Models")
    plt.ylabel("Difference from Mean")
    plt.xlabel("Metrics")
    plt.xticks(x + width, metrics)
    plt.axhline(0, color='grey', linestyle='--')
    plt.legend(title="Model")
    
    # Annotate each bar
    for i, (model, color) in enumerate(zip(metrics_dict.keys(), ['skyblue', 'lightgreen', 'orange', 'pink'])):
        model_diffs = differences[model]
        for j, diff in enumerate(model_diffs):
            plt.text(x[j] + i * width, diff + (0.001 if diff > 0 else -0.001), f"{diff:.3f}",
                     ha='center', va='bottom' if diff > 0 else 'top')
    
    plt.tight_layout()
    plt.show()

#############################################
# 4) Main Workflow
#############################################
if __name__ == "__main__":
    print("\n Plotting Malignant Metrics Across Models:")
    plot_malignant_metrics(malignant_metrics)
    
    print("\n Plotting Differences from Mean for Malignant Metrics Across Models:")
    plot_malignant_differences(malignant_metrics)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define confusion matrices for all models
confusion_matrices = {
    "DenseNet121": [[531, 5], [11, 451]],
    "ResNet50": [[505, 31], [16, 446]],
    "EfficientNet-B4": [[514, 22], [26, 436]],
    "CustomCNN": [[463, 73], [16, 446]],
    "XGBoost": [[408, 20], [18, 352]],
    "CatBoost": [[412, 16], [20, 350]],
    "LightGBM": [[412, 16], [19, 350]],
    "Ensemble": [[408, 20], [16, 354]]
}

# Titles for clarity
titles = {
    "DenseNet121": "DenseNet121",
    "ResNet50": "ResNet50",
    "EfficientNet-B4": "EfficientNet-B4",
    "CustomCNN": "CustomCNN",
    "XGBoost": "XGBoost",
    "CatBoost": "CatBoost",
    "LightGBM": "LightGBM",
    "Ensemble": "Ensemble Model"
}

#############################################
# 1) Plot Confusion Matrices Side-by-Side
#############################################
def plot_confusion_matrices_side_by_side(conf_matrices, titles):
    """
    Plot all confusion matrices side-by-side for easy comparison.
    """
    num_models = len(conf_matrices)
    cols = 4  # Number of columns per row
    rows = (num_models + cols - 1) // cols  # Calculate number of rows dynamically
    
    fig, axes = plt.subplots(rows, cols, figsize=(20, 5 * rows))
    axes = axes.flatten()  # Flatten axes for easy iteration
    
    for i, (model, cm) in enumerate(conf_matrices.items()):
        ax = axes[i]
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax)
        ax.set_title(titles[model])
        ax.set_xlabel("Predicted Labels")
        ax.set_ylabel("True Labels")
    
    # Hide any unused axes if the grid is not perfectly filled
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.tight_layout()
    plt.show()

#############################################
# 2) Focus on Malignant Class Performance
#############################################
def plot_malignant_focus(conf_matrices, titles):
    """
    Focus only on the performance of the Malignant class (True Positive, False Negative).
    """
    malignant_data = {}
    for model, cm in conf_matrices.items():
        tp = cm[1][1]  # True Positive (Malignant predicted as Malignant)
        fn = cm[1][0]  # False Negative (Malignant predicted as Benign)
        malignant_data[model] = {"True Positive": tp, "False Negative": fn}
    
    models = list(malignant_data.keys())
    true_positive = [malignant_data[m]["True Positive"] for m in models]
    false_negative = [malignant_data[m]["False Negative"] for m in models]
    
    x = range(len(models))
    width = 0.4
    
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.bar(x, true_positive, width, label='True Positive', color='seagreen')
    ax.bar([i + width for i in x], false_negative, width, label='False Negative', color='salmon')
    
    ax.set_ylabel('Number of Samples')
    ax.set_title('Malignant Class: True Positive vs False Negative Across Models')
    ax.set_xticks([i + width / 2 for i in x])
    ax.set_xticklabels(models, rotation=45, ha='right')
    ax.legend()
    
    # Annotate each bar
    for i in range(len(models)):
        ax.text(i, true_positive[i] + 1, str(true_positive[i]), ha='center', va='bottom')
        ax.text(i + width, false_negative[i] + 1, str(false_negative[i]), ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()

#############################################
# 3) Focus on True Positive Rate (Malignant)
#############################################
def plot_malignant_tpr(conf_matrices, titles):
    """
    Plot the True Positive Rate (TPR) for Malignant class across all models.
    """
    tpr_values = {}
    for model, cm in conf_matrices.items():
        tp = cm[1][1]  # True Positive
        fn = cm[1][0]  # False Negative
        tpr = tp / (tp + fn) if (tp + fn) > 0 else 0
        tpr_values[model] = tpr
    
    models = list(tpr_values.keys())
    tpr_scores = list(tpr_values.values())
    
    fig, ax = plt.subplots(figsize=(16, 8))
    bars = ax.bar(models, tpr_scores, color='royalblue')
    
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Malignant Class: True Positive Rate Across Models')
    ax.set_ylim(0.9, 1.0)  # Compact Y-axis for better visualization
    
    # Annotate each bar
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset
                    textcoords="offset points",
                    ha='center', va='bottom')
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

#############################################
# 4) Main Workflow
#############################################
if __name__ == "__main__":
    print("\n Plotting Confusion Matrices Across Models:")
    plot_confusion_matrices_side_by_side(confusion_matrices, titles)
    
    print("\n Focusing on Malignant Class Performance (TP & FN):")
    plot_malignant_focus(confusion_matrices, titles)
    
    print("\n Plotting True Positive Rate (TPR) for Malignant Class Across Models:")
    plot_malignant_tpr(confusion_matrices, titles)
```
